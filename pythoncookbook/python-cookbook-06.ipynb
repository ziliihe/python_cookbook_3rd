```python
# 6.5 将 python 字典中的数据转换为 XML
from lxml import etree

def dict_to_xml(tag, d):
    """
    Turn a simple dict of key/value pairs into XML
    """
    elem = etree.Element(tag)
    for key, val in d.items():
        child = etree.Element(key)
        child.text = str(val)
        elem.append(child)
    return elem

a = {'name': 'GOOG', 'shares': 100, 'price': 491.1}
b = dict_to_xml('stock', a)
print(b)
print(b.tag)
b.set('_id', '1234')
print(etree.tostring(b))

from xml.sax.saxutils import escape, unescape
c = escape('<spam>')
d = unescape(c)
print(c)
print(d)
```

```python
# 6.7 使用命名空间来解析 XML 文档
class XMLNamespaces:
    def __init__(self, **kwargs):
        self.namespaces = {}
        for name, uri in kwargs.items():
            self.register(self, name, uri)
    def register(self, name, uri):
        self.namesapces[name] = f"{{uri}}"

    def __call__(self, path):
        return path.format_map(self.namespaces)
```

```python
# 6.11 读写二进制结构的数组
from struct import Struct

def write_records(records, format, f):
    """
    write a sequence of tuples to a binary file of structures
    """
    record_struct = Struct(format)
    for r in records:
        f.write(record_struct.pack(*r))

records = [
    (1, 2.3, 4.5),
    (6, 7.8, 9.0),
    (12, 13.4, 56.7)
]
# <idd: <表示使用小端字节顺序，i表示一个整数类型，d表示一个双精度浮点数
with open('resource/o.b', 'wb') as f:
    write_records(records, '<idd', f)

def read_records(format, f):
    record_struct = Struct(format)
    # 迭代器会重复调用提供的第一个函数，直到返回一个指定值为止
    chunks = iter(lambda: f.read(record_struct.size), b'')
    return (record_struct.unpack(chunk) for chunk in chunks)

with open('resource/o.b', 'rb') as f:
    a = read_records('<idd', f)
    print(list(a))


def unpack_records(format, data):
    record_struct = Struct(format)
    # unpack_from 不会创建临时对象，不会执行内存拷贝，通过字节串和偏移量进行读取
    return (record_struct.unpack_from(data, offset) for offset in range(0, len(data), record_struct.size))

with open('resource/o.b', 'rb') as f:
    data = f.read()

for i in range(0, len(data), Struct('<idd').size):
    print(i, end=' ')

b = unpack_records('<idd', data)
print()
print(list(b))

c = Struct('<idd')
print(c.size)
d = c.pack(1, 2.1, 3.3)
print(d)
e = c.unpack(d)
print(e)
```

```python
# 6.12 读取嵌套型和大小可的二进制结构
polys = [
    [(1.0, 2.4), (3.5, 4.0), (2.4, 1.4)],
    [(7.0, 1.2), (4.1, 5.3), (0.5, 7.1), (0.5, 0.9)],
    [(3.4, 5.3), (1.2, 0.5), (5.4, 9.1)]
]

import struct
import itertools

def write_polys(filename, polys):
    flattened = list(itertools.chain(*polys))
    min_x = min(x for x, y in flattened)
    max_x = max(x for x, y in flattened)
    min_y = min(y for x, y in flattened)
    max_y = max(y for x, y in flattened)
    
    with open(filename, 'wb') as f:
        f.write(struct.pack('<iddddi', 0x1234, min_x, min_y, max_x, max_y, len(polys)))

        for poly in polys:
            size = len(poly) * struct.calcsize('<dd')
            f.write(struct.pack('<i', size+4))
            for pt in poly:
                f.write(struct.pack('<dd', *pt))
write_polys('resource/p.bin', polys)

# def read_polys(filename, polys):
    
```
